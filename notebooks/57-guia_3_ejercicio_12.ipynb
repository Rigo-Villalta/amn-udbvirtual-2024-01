12. Un tanque de gasolina cilíndrico está colocado con su eje en posición horizontal. Obtenga la fuerza del fluido sobre una de las paredes del tanque si éste está medio lleno, considere el diámetro igual a 8 pies y que la gasolina pesa 42 libras/pie<sup>3</sup>. Emplee la regla compuesta del trapecio con n=8. Además, obtenga el valor exacto y el error de aproximación.

Deducción de fórmula:



```python
# Solución en python

from sympy import * 

y = symbols("y")

# Defino mi función simbólica

f = 84 * (4-y) * sqrt(16 - (y-4)**2)

a = 0
b = 4
n = 8

# Hago una función que calcula la fórmula compuesta del trapecio
def regla_compuesta_del_trapecio(f, a, b, n):
    # Defino mi h dentro de la función
    h = (b-a) / n
    # La suma interna es el segundo factor de la fórmula, la inicio con f(x_0)
    suma_interna = f.subs(y, a)
    # Aquí hago la sumatoria desde f(y_1) hasta f(y_(n-1)) 
    for i in range(1, n):
        suma_interna += 2 * f.subs(y, a + i*h)
    # Sumo f(x_n)
    suma_interna += f.subs(y, b)
    return (h/2) * suma_interna

valor_aproximado = regla_compuesta_del_trapecio(f, a, b, n)
print((
    f"El valor aproximado de la fuerza del"
    f" fluido sobre una de las paredes es: {valor_aproximado} Libras fuerza"))
```

```python
valor_exacto = float(integrate(f, (y, a, b)))
print((
    f"El valor exacto de la fuerza del fluido sobre una de las paredes es:"
    f" {valor_exacto} Libras fuerza")
)
```

```python
error = abs(valor_exacto - valor_aproximado)

error_relativo = round((error/valor_exacto) * 100, 2)

print("El error absoluto es de:", error)
print(f"El error relativo es de: {error_relativo}%")
```
