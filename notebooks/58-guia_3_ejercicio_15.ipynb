15 . Aproxime el volumen del sólido de revolución generado al rotar la región acotada por: y = 4 – x<sup>2</sup>, y = x<sup>2</sup> – 2x + 4, alrededor de la recta y = 7. Emplee la regla compuesta de Simpson con n=10. Además, obtenga el valor exacto y el error de aproximación.

Planteo mi integral a partir de la gráfica:


```python
# Desarrollo en Python

from sympy import *

x = symbols("x")

# Declaro la función simbólica
f = pi * ( (3 - x**2 + 2*x)**2  - (3 + x**2)**2 )
a = 0
b = 1
n = 10


# Defino una función que resuelve la regla compuesta de Simpson
def regla_compuesta_del_simpson(f, a, b, n):
    # Defino mi h dentro de la función
    h = (b-a) / n
    # La suma interna es el factor derecho de la fórmula, la inicio con f(x_0)
    suma_interna = f.subs(x, a)
    # Aquí hago las sumatorias desde f(x_1) hasta f(x_(n-1)) y hasta f(x_(n-2))
    # No es necesario el n-2 si n es par así que hago un solo for
    # Recordando que en Python los índices son iguales que en la fórmula
    for i in range(1, n):
        if i%2 == 0:
            suma_interna += 2 * f.subs(x, a + i*h)
        else:
            suma_interna += 4 * f.subs(x, a + i*h)
    # Sumo f(x_n)
    suma_interna += f.subs(x, b)
    return (h/3) * suma_interna

volumen_aproximado = float(regla_compuesta_del_simpson(f, a, b, n))

print(
    f"El volumen aproximado del sólido de revolución es de:"
    f"{volumen_aproximado} unidades cúbicas"
)
```

```python
volumen_exacto = float(integrate(f, (x, a, b)))
print(
    f"El volumen exacto del sólido de revolución es de:"
    f"{volumen_exacto} unidades cúbicas"
)
```

```python
error_absoluto = abs(volumen_exacto-volumen_aproximado)

print(
    f"El error es: {error_absoluto}"
)
```
