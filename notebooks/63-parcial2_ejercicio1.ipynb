En el clculo anterior el origen est en el medio de la base.

Teniendo mi integral y mis lmites paso a resolverlo en Pyhton

```python
from sympy import *

x = symbols("x")

a = 0
b= 4
x_0 = a
x_5 = b
h = (b-a)/ 5
x_1 = a + h
x_2 = a + 2*h
x_3 = a + 3*h
x_4 = a + 4*h

f_de_x = 9810*(4-x)*(x+8)


valor_aproximado = ((5 * h) / 288) * (
    (19 * f_de_x.subs(x, x_0)) +  (75 * f_de_x.subs(x, x_1)) + (50 * f_de_x.subs(x, x_2)) + 
    (50 * f_de_x.subs(x, x_3)) + (75 * f_de_x.subs(x, x_4)) + (19 * f_de_x.subs(x, x_5)))

print("El valor aproximado es:", round(valor_aproximado, 8))

valor_exacto = integrate(f_de_x, (x, a, b))
print("El valor exacto es:", round(valor_exacto, 8))

error = abs(valor_exacto - valor_aproximado)
print("El error es de:", float(error))
```
