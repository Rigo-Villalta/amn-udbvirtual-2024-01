3 - Aproxime 𝑓(1.3675) a partir de los siguientes datos

| x | 1.27 | 1.29 | 1.31 | 1.33 | 1.35 | 1.7 |
|---|-----|-----|-----|-----|-----|-----|
| f(x) | 13.270567 | 13.781763 | 14.307413 | 14.847887 | 15.403567 | 15.974842 |

Además, calcule el valor exacto y el error de aproximación si la función es:

$$ f(x) = 3 x e^x - cos(x)  $$

Se debe mostrar:

- Las funciones L construidas.
- El polinomio de interpolación
- El valor aproximado de la función

```python
from math import e, cos

from metodos_interpolacion import lagrange
from utils import  imprimir_tabla

data_x = [1.27, 1.29, 1.31, 1.33, 1.35, 1.37]
data_y = [13.270567, 13.781763, 14.307413, 14.847887, 15.403567, 15.974842]

def fdx(x):
    return 3*x*e**x - cos(x)

resultado = lagrange(data_x, data_y, 1.3675, 15)
imprimir_tabla(resultado)

resultado_aproximado = 15.9025664007683
print(f"El resultado aproximado de 1.3675 es de: {resultado_aproximado}")
resultado_exacto = round(fdx(1.3675), 15)
print(f"El resultado exacto de 1.3675 es de: {resultado_exacto}")
error = abs(resultado_aproximado - resultado_exacto)
print(f"El error es de {error:.1E}")
```
